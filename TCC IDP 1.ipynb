```python
import pandas as pd
import os
from IPython.display import display, Image
import ipywidgets as widgets
```

```python
pasta = "D:\\Github\\TCC IDP"
caminho_imagem = os.path.join(pasta, "20220913_migracao_interna.jpg")
caminho_saldo = os.path.join(pasta, "saldo_migratorio_estados.csv")
caminho_pop = os.path.join(pasta, "pop_migrantes.csv")
```

```python
print("TCC - IDP: Fluxo migratório Brasileiro")
print("Brasil em Movimento: dados do Censo 2022 revelam novos padrões migratórios\n")

display(Image(filename=caminho_imagem, width=600))

print(
    "A migração interna no Brasil é um fenômeno dinâmico e fundamental para compreender "
    "as transformações sociais e econômicas do país. O Censo 2022 oferece uma oportunidade "
    "de análise com os últimos dados divulgados dos movimentos populacionais após mais de uma década.\n"
)
```

```python
try:
    df_saldo = pd.read_csv(caminho_saldo)
    df_saldo['saldo_migratorio'] = pd.to_numeric(df_saldo['saldo_migratorio'], errors='coerce')
except FileNotFoundError:
    print(f"Arquivo não encontrado em: {caminho_saldo}")
except Exception as e:
    print(f"Ocorreu um erro ao carregar a tabela de saldo migratório: {e}")
```

```python
def mostrar_saldo_callback(b):
    display(df_saldo)

btn_saldo = widgets.Button(description="Confira aqui o saldo migratório brasileiro")
btn_saldo.on_click(mostrar_saldo_callback)
display(btn_saldo)
```

```python
try:
    df_pop = pd.read_csv(caminho_pop)
except FileNotFoundError:
    print(f"Arquivo não encontrado em: {caminho_pop}")
except Exception as e:
    print(f"Ocorreu um erro ao carregar a tabela de migrantes: {e}")
```

```python
def mostrar_pop_callback(b):
    display(df_pop)

btn_pop = widgets.Button(description="Confira aqui os estados que mais receberam migrantes (e de qual estado eles vieram)")
btn_pop.on_click(mostrar_pop_callback)
display(btn_pop)
```
